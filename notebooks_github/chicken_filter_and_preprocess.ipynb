# Notes on QC  
Initially I filtered cells based on QC variables.  
Then I reassessed the filtered cells and ran scrublet for doublet simulation.

I will filter both with cutoffs and clusters. Sometimes it's difficult to design cutoffs that capture all the bad cells.  
Clustering will often group them together well. If when assessing a sample I find a cluster of what I percieve as entirely bad cells, I will remove the cluster.

```python
import logging
logging.getLogger('matplotlib.font_manager').setLevel(logging.ERROR)
import scanpy as sc
import scvelo as scv
import scvi
import seaborn as sns
import plotly.express as px
import numpy as np
from dash import Dash, dcc, html, Input, Output

import pandas as pd

import warnings
import logging
import os
import sys
import time
import gc
os.environ['R_HOME'] = sys.exec_prefix+"/lib/R/"

# Plotting
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.backends.backend_pdf import PdfPages
from matplotlib.colors import LinearSegmentedColormap, ListedColormap
from matplotlib.lines import Line2D 

import logging
logger = logging.getLogger()
logger.setLevel(logging.CRITICAL)

from copy import copy
reds = copy(mpl.cm.Reds)
reds.set_under("lightgray")

project_directory = '/Cranio_Lab/Louk_Seton/4_species_project'
os.chdir(os.path.expanduser("~")+project_directory)
```

```python
##TODO

######chicken timeeee######
start_time=time.strftime("%Y_%m_%d-%I_%M_%S_%p")
print('start time:',start_time)

sample_names = ['CkD1','CkD2','CkD3','CkD4','CkD5',] #specify the sample names
species = 'chicken' #specify the species
genome = 'GRCg7b' #specify the genome
output_prefix = 'cellranger_related/cellranger_output/sc_rnaseq/' #specify the location of the cellranger output

mt_genes = ['ND1','ND2','ND3','ND4','ND4L','ND5','ND6','COX1','COX2','COX3','ATP6','ATP8','CYTB']
rb_genes = ('RPS','RPL')
hb_genes = ['HBBA','HBBR','HBAD','HBA1','HBZ','HBE','HBE1',]


##plotting vars
output_dir = 'figures_publish/'+species+'/'+genome+'/QC_plots/'
!mkdir -p {output_dir}
genes_to_plot = ['SOX10',]
dot_size = .6
linewidth = 0
scatter_color = 'pct_counts_hb'
scatter_cmap = 'coolwarm'
y1 = 'pct_counts_mt' #these are for the QC scatter plots
y2 = 'pct_counts_rb'
x1 = 'n_genes_by_counts'
x2 = 'total_counts'
##################################################################################################################
cellranger_locs = []
velocyto_locs = []

for sample in sample_names: #simple loop to create a list with the location of the required files
    cellranger_locs = cellranger_locs + [output_prefix+species+'/'+genome+'/'+sample+'/outs/filtered_feature_bc_matrix/']
    velocyto_locs = velocyto_locs + [output_prefix+species+'/'+genome+'/'+sample+'/velocyto/'+sample+'.loom']

warnings.filterwarnings('ignore')
logging.getLogger().setLevel(logging.CRITICAL)

adata_files={}
print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Reading file(s)')
for i,j,k in zip(sample_names, cellranger_locs, velocyto_locs):
    ##Read and merge 10x and velocyto output, preserving gene names and ensemblid
    print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Sample ',i)
    adata_cellr = sc.read_10x_mtx(j)
    print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Loaded 10x mtx')
    adata_cellr.var['gene_name'] = adata_cellr.var.index
    adata_cellr.var.index = adata_cellr.var['gene_ids']
    adata_cellr.obs.index = adata_cellr.obs.index.str[:-2]
   
    adata_veloc = sc.read(k)
    print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Loaded velocyto loom')
    adata_veloc.obs.index = adata_veloc.obs.index.str.split(':').str[1]
    adata_veloc.obs.index = adata_veloc.obs.index.str[:-1]
    adata_veloc.var.index = adata_veloc.var['Accession']
    print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Merging data')
    adata_merge = scv.utils.merge(adata_cellr, adata_veloc)
    adata_files[i]=adata_merge
    
    adata_files[i].obs['sample']=i
    adata_files[i].obs['ref_genome']=genome
    adata_files[i].obs['barcode']=adata_files[i].obs.index
    adata_files[i].obs.index=adata_files[i].obs['barcode']+'.'+adata_files[i].obs['sample']

    adata_files[i].var.index = list(adata_files[i].var['gene_name'])#set gene_names as index

    ##Calculate QC values
    print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Calculating QC values for cells')
    adata_files[i].var['mt'] = adata_files[i].var['gene_name'].isin(mt_genes)  # annotate the group of mitochondrial genes as 'mt'
    adata_files[i].var['rb'] = adata_files[i].var['gene_name'].str.startswith(rb_genes)  # annotate the group of ribosomal proteins as 'rb'

    adata_files[i].var['hb'] = adata_files[i].var['gene_name'].isin(hb_genes) #check this
    adata_files[i].var_names_make_unique()
    adata_files[i].obs_names_make_unique()
    sc.pp.calculate_qc_metrics(adata_files[i], qc_vars=["mt", "rb", "hb"], inplace=True, log1p=True)
    #clip to meaningful numbers, better for comparisons
    adata_files[i].obs['pct_counts_mt'] = adata_files[i].obs['pct_counts_mt'].clip(upper = 20)
   # adata_files[i].obs['pct_counts_hb'] = adata_files[i].obs['pct_counts_hb'].clip(upper = 20)
    #adata_files[i].obs['pct_counts_rb'] = adata_files[i].obs['pct_counts_rb'].clip(upper = 40)
    adata_files[i].obs['total_counts'] = adata_files[i].obs['total_counts'].clip(upper = 50000)

    ##calculate doublet score
    print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Running scrublet for doublet detection')
    sc.pp.scrublet(adata_files[i])

    print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Initial preprocess and clustering')
    ##standard normalization
    adata_files[i].layers["original_counts"] = adata_files[i].X.copy() #store original counts
    sc.pp.normalize_total(adata_files[i]) # Normalizing to median total counts
    sc.pp.log1p(adata_files[i]) # Logarithmize the data
    adata_files[i].layers["normalized_counts"] = adata_files[i].X.copy()

    ##highly variable genes
    sc.pp.highly_variable_genes(adata_files[i], n_top_genes=2000,)

    ##dimensionality reduction and clustering
    sc.tl.pca(adata_files[i])
    sc.pp.neighbors(adata_files[i])
    sc.tl.umap(adata_files[i])
    sc.tl.leiden(adata_files[i],)

    ##cell cycle scoring
    print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Prepare cell cycle scoring')
    cell_cycle_genes = [x.strip() for x in open('required_files/chicken_cell_cycle.txt')]
    s_genes = cell_cycle_genes[:43]
    s_genes = [x for x in s_genes if x in list(adata_files[i].var['gene_name'])]
    g2m_genes = cell_cycle_genes[43:]
    g2m_genes = [x for x in g2m_genes if x in list(adata_files[i].var['gene_name'])]
    print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Cell cycle scoring')
    sc.tl.score_genes_cell_cycle(adata_files[i], s_genes=s_genes, g2m_genes=g2m_genes)

    
    #plot pdfs of qc
    print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Plotting QC and saving to PDF')
    with PdfPages(output_dir+i+'_QC.pdf') as pdf:
        mpl.rcParams['axes.grid'] = False
        ax = sc.pl.violin(adata_files[i],["n_genes_by_counts", "total_counts", "pct_counts_mt","pct_counts_hb", "pct_counts_rb",'doublet_score'],
                          stripplot=False, multi_panel=True,show=False,
                          height=4, aspect=.8,col_wrap = 3)
        plt.suptitle(i, x = 0.1)
        pdf.savefig()  # saves the current figure into a pdf page
        plt.close() # close the plot and page

        ###Scatterplots
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2,layout='constrained',figsize=(10,8))
        s=dot_size
        linewidth=linewidth
        cmap=scatter_cmap
        c = scatter_color
        ax1.scatter(data=adata_files[i].obs, x=x1,y=y1,
                    c = c,s = s,linewidth=linewidth, cmap = cmap,rasterized=True)
        ax1.set_xlim(adata_files[i].obs[x1].quantile(0) ,adata_files[i].obs[x1].quantile(0.95))
        ax1.set_xlabel(x1) 
        ax1.set_ylabel(y1)
        ax2.scatter(data=adata_files[i].obs, x=x2,y=y1,
                    c = c,s = s,linewidth=linewidth, cmap = cmap,rasterized=True)
        ax2.set_xlim(adata_files[i].obs[x2].quantile(0) ,adata_files[i].obs[x2].quantile(0.95))
        ax2.set_xlabel(x2) 
        ax2.set_ylabel(y1)
        ax3.scatter(data=adata_files[i].obs, x=x1,y=y2,
                    c = c,s = s,linewidth=linewidth, cmap = cmap,rasterized=True)
        ax3.set_xlim(adata_files[i].obs[x1].quantile(0) ,adata_files[i].obs[x1].quantile(0.95))
        ax3.set_xlabel(x1) 
        ax3.set_ylabel(y2)
        ax4.scatter(data=adata_files[i].obs, x=x2,y=y2,
                    c = c,s = s,linewidth=linewidth, cmap = cmap,rasterized=True)
        ax4.set_xlim(adata_files[i].obs[x2].quantile(0) ,adata_files[i].obs[x2].quantile(0.95))
        ax4.set_xlabel(x2) 
        ax4.set_ylabel(y2)


        fig.colorbar(mpl.cm.ScalarMappable(norm=mpl.colors.Normalize(0, max(adata_files[i].obs[c])), cmap=cmap),
                     ax=ax1, orientation='vertical', label=c)
        pdf.savefig(dpi=150)
        plt.close()

        ##nice jointplot shows good cells
        mpl.rcParams['axes.grid'] = False
        sns_plot = sns.jointplot(data=adata_files[i].obs,
                                 x="log1p_total_counts",
                                 y="log1p_n_genes_by_counts",
                                 kind="hex",
                                 marginal_kws=dict(bins=100,),
                                 joint_kws = dict(gridsize=50,),
                                )
        sns_plot.fig.suptitle(i,x=0.05)
        pdf.savefig(dpi=150)
        plt.close()

        ##custom hvg plot
        fig, ((ax1), (ax2)) = plt.subplots(1,2,layout='constrained',figsize = (8,4))
        s=dot_size
        linewidth=linewidth
        cmap = ListedColormap(['grey','black'])
        c = 'highly_variable'
        scatter = ax1.scatter(data=adata_files[i].var, x='means',y="dispersions",
                              c=c,s = s,cmap=cmap,linewidth=linewidth,rasterized=True)
        ax1.set_xlabel('mean expression') 
        ax1.set_ylabel('dispersion') 
        ax1.legend(scatter.legend_elements()[0][0:2],['other genes','highly variable'],
                   loc="upper right",)
        scatter = ax2.scatter(data=adata_files[i].var, x='means',y="dispersions_norm",
                              c=c,s = s,cmap=cmap,linewidth=linewidth,rasterized=True)
        ax2.legend(scatter.legend_elements()[0][0:2],['other genes','highly variable'],
                   loc="upper right",)
        ax2.set_xlabel('mean expression') 
        ax2.set_ylabel('dispersions_norm')
        pdf.savefig(dpi=150)
        plt.close()

        ##PCA variance plot
        sc.pl.pca_variance_ratio(adata_files[i], log=True, show=False)
        pdf.savefig()
        plt.close()
        
        ##umaps
        plt.rcParams['figure.figsize'] = [5,4]
        ax = sc.pl.umap(adata_files[i],
                        color=['doublet_score','leiden','total_counts','pct_counts_mt','pct_counts_rb','pct_counts_hb','phase',]+genes_to_plot,
                        show = False,ncols = 2)
        for p in ax:
            p.set_rasterized(True)
        pdf.savefig(dpi=150)
        plt.close()

        ##Most highly expressed genes
        plt.rcParams['figure.figsize'] = [8,4]
        adata_files[i].X = adata_files[i].layers["original_counts"].copy() #set raw counts to X because the layer argument is broken
        ax = sc.pl.highest_expr_genes(adata_files[i], 
                                      gene_symbols = 'gene_name',
                                      n_top=20,log=True,show=False)
        ax.set_rasterized(True)
        pdf.savefig()
        plt.close()
        ##return normalised counts to X
        adata_files[i].X = adata_files[i].layers["normalized_counts"].copy()

        ##doublet scatterplots
        ###Scatterplots
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2,layout='constrained',figsize=(10,8))
        s=dot_size
        linewidth=linewidth
        cmap=scatter_cmap
        c = scatter_color
        ax1.scatter(data=adata_files[i].obs, x=x1,y='doublet_score',
                    c = c,s = s,linewidth=linewidth, cmap = cmap,rasterized=True)
        ax1.set_xlabel(x1) 
        ax4.set_ylabel('doublet_score')
        ax2.scatter(data=adata_files[i].obs, x=x2,y='doublet_score',
                    c = c,s = s,linewidth=linewidth, cmap = cmap,rasterized=True)
        ax2.set_xlabel(x2) 
        ax4.set_ylabel('doublet_score')
        ax3.scatter(data=adata_files[i].obs, x=y1,y='doublet_score',
                    c = c,s = s,linewidth=linewidth, cmap = cmap,rasterized=True)
        ax3.set_xlabel(y1) 
        ax4.set_ylabel('doublet_score')
        ax4.scatter(data=adata_files[i].obs, x=y2,y='doublet_score',
                    c = c,s = s,linewidth=linewidth, cmap = cmap,rasterized=True)
        ax4.set_xlabel(y2) 
        ax4.set_ylabel('doublet_score')
        fig.colorbar(mpl.cm.ScalarMappable(norm=mpl.colors.Normalize(0, max(adata_files[i].obs[c])), cmap=cmap),
                     ax=ax1, orientation='vertical', label=c)
        pdf.savefig(dpi=150)
        plt.close()
        
        #boxplot of doublet scores
        ax = sns.boxplot(data=adata_files[i].obs, x="leiden", y="doublet_score",
            palette = adata_files[i].uns['leiden_colors'], 
                 fliersize=0)
        cnum = 0
        for line in ax.lines:
            if line.get_linestyle() == 'None':
                sns.swarmplot(x=[int(b) for b in line.get_xdata()], y=line.get_ydata(),
                              size = 1.5, color = adata_files[i].uns['leiden_colors'][cnum], rasterized = True)
                cnum = cnum+1
        pdf.savefig(dpi=150)
        plt.close()

    ##saving h5ad
    print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Saving data to h5ad file')
    !mkdir -p {'h5ad_files/'+species+'/'+genome}
    adata_files[i].write('h5ad_files/'+species+'/'+genome+'/'+i+'_before_filtering.h5ad')
    

    ##remove used variables from memory
    del adata_cellr
    del adata_veloc
    del adata_merge
    gc.collect()
    print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Loaded files for ',i)
print(time.strftime("%Y_%m_%d-%I_%M_%S_%p"),'Finished Reading file(s)')
```
